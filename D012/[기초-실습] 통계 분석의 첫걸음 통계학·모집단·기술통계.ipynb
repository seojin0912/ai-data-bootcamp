# 문제 2. 평균 vs 중앙값 비교

**📘 문제**

- A지역과 B지역의 연봉 데이터를 비교해 봅시다.
- B 지역에는 일부 고소득자가 포함되어 있습니다.

```python
# 데이터 생성
np.random.seed(2025)

# A 지역: 평균 300, 표준편차 50
region_a = np.random.normal(300, 50, 100)

# B 지역: 일반 소득자 + 극단적인 고소득자 포함
region_b = np.concatenate([
    np.random.normal(280, 40, 95),    # 일반 소득자 95명
    np.random.normal(1500, 200, 5)    # 고소득자 5명
])

df2 = pd.DataFrame({
    'region': ['A'] * 100 + ['B'] * 100,
    'salary': np.concatenate([region_a, region_b])
})
```

**📌 아래를 수행해 보세요:**

- 각 지역의 평균 연봉과 중앙값을 구해 봅시다.
- 지역별 박스플롯(Boxplot)을 그려서 분포의 차이를 시각적으로 확인해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 2] 여기에 코드를 작성해주세요.

summary = df2.groupby('region')['salary'].agg(['mean', 'median'])
print(summary)
df2.boxplot(column='salary', by='region')
plt.title('지역별 연봉 분포')
plt.suptitle('')  # pandas가 자동으로 붙이는 기본 제목 제거
plt.xlabel('지역')
plt.ylabel('연봉')
plt.show()
```

# 문제 3. 이상값과 분산 비교

**📘 문제**

- 1반과 2반의 시험 점수 분포를 비교해 봅시다.
- 2반에는 극단적인 이상값(예: 부정행위로 0점)이 포함되어 있습니다.

```python
# 데이터 생성
np.random.seed(2025)  # 재현 가능성을 위해 시드 고정

# 1반: 평균 75, 표준편차 8
class1 = np.random.normal(75, 8, 30)

# 2반: 일반 학생 + 이상값(0점, 130점)
class2 = np.concatenate([
    np.random.normal(75, 15, 28),  # 일반 학생
    [0, 130]                       # 극단적인 이상값
])

df3 = pd.DataFrame({
    'class': ['1반'] * 30 + ['2반'] * 30,
    'score': np.concatenate([class1, class2])
})
```

**📌 아래를 수행해 보세요:**

1. 각 반의 평균, 표준편차, 최소/최대값을 계산해 봅시다.
2. 박스플롯을 사용하여 두 반의 분포 차이를 시각적으로 비교해 봅시다.
3. 이상값이 통계 지표에 어떤 영향을 주는지 생각해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 3] 여기에 코드를 작성해주세요.

summary = df3.groupby('class')['score'].agg(['mean', 'std', 'min', 'max'])
print(summary)

df3.boxplot(column='score', by='class')
plt.title('class-score')
plt.suptitle('')
plt.xlabel('class')
plt.ylabel('score')
plt.show()

# 이상값이 통계 지표에 어떤 영향을 주는지 생각해 봅시다
# 평균과 표준편차는 크게 영향을 받을 것이다. 
# 그러나 중앙값은 크게 영향을 받지 않을 것이고, 
# 최소-최대값은 그 자체가 이상치가 될 것이다.
```

# 문제 4. IQR 이상값 제거

**📘 문제**

- 앞의 문제에서 2반에는 **극단적인 이상값(예: 매우 낮은 점수)**이 포함되어 있어

- 이번에는 **IQR(사분위 범위)**을 기준으로 이상값을 제거한 뒤, 이상값 제거 전후의 평균을 비교해 봅시다

<br>

**📌 아래를 수행해 보세요:**

- 이상값 제거 전과 제거 후의 평균을 각각 계산해 봅시다.
- 이상값 제거가 평균에 어떤 영향을 주는지 설명해 봅시다.
- 필요하다면 중앙값과도 비교해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 4] 여기에 코드를 작성해주세요.

# 2반 데이터만 추출
class2_scores = df3[df3['class'] == '2반']['score']

# IQR 계산
Q1 = class2_scores.quantile(0.25)
Q3 = class2_scores.quantile(0.75)
IQR = Q3 - Q1

# 이상값 경계 설정 (일반적으로 1.5 * IQR 기준)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
print(f"이상값 경계: {lower_bound:.2f} ~ {upper_bound:.2f}")

# 이상값 판별
outliers = class2_scores[(class2_scores < lower_bound) | (class2_scores > upper_bound)]
print(f"\n이상값: {outliers.values}")

# 이상값 제거
class2_clean = class2_scores[(class2_scores >= lower_bound) & (class2_scores <= upper_bound)]

#제거 전/후 평균 비교
mean_before = class2_scores.mean()
mean_after = class2_clean.mean()

print(f"제거 전 평균: {mean_before:.2f}")
print(f"제거 후 평균: {mean_after:.2f}")
print(f"차이: {mean_before - mean_after:.2f}")

# 이상값을 제거하면 평균이 
# "전형적인 학생들의 실제 수준"에 더 가까운 값으로 이동


# 중앙값 비교 
median_before = class2_scores.median()
median_after = class2_clean.median()

print(f"제거 전 중앙값: {median_before:.2f}")
print(f"제거 후 중앙값: {median_after:.2f}")
```

# 문제 9. 정규분포와 표준편차

**📘 문제**

- 정규분포와 표준편차의 개념을 실제 데이터에 적용해 보겠습니다.
- 정규분포를 따르는 점수 데이터에서 다음을 확인해 봅시다:
  - 평균 ±1σ, 평균 ±2σ 범위에 속한 값들의 비율은 실제로 얼마나 되는가?
  - 이 결과가 이론적인 68-95-99 법칙과 얼마나 유사한지 비교해 봅시다.

<br>

**📌 아래를 수행해 보세요:**

1. 평균이 70, 표준편차가 10인 정규분포 데이터를 1000개 생성해 봅시다.
2. 평균을 중심으로 ±1σ, ±2σ 범위에 포함되는 데이터의 비율을 계산해 봅시다.
3. 결과를 바탕으로 68-95-99 법칙과 비교해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 9] 여기에 코드를 작성해주세요.

np.random.seed(42)
data = np.random.normal(70, 10, 1000)

mean = data.mean()
std = data.std()
print(f"실제 평균: {mean:.2f}")   # 70.19
print(f"실제 표준편차: {std:.2f}") # 9.79

for k in [1, 2, 3]:
    lower = mean - k*std
    upper = mean + k*std
    within = ((data >= lower) & (data <= upper)).sum()
    pct = within / len(data) * 100
    print(f"±{k}σ 범위: {pct:.2f}%")

# 이론값 (68-95-99.7 법칙)/ 실제 측정값/차이
# ±1σ → 68.27%	68.60%	0.33%p
# ±2σ → 95.45%	95.60%	0.15%p
# ±3σ → 99.73%	99.70%	0.03%p

# 일치
```

**📘 문제**

- 두 그룹의 전환율 차이를 계산하고 해석해봅시다.

```python
# 데이터 생성
np.random.seed(606)
group_a = np.random.binomial(1, 0.12, 1000)
group_b = np.random.binomial(1, 0.15, 1000)
```

**📌 아래를 수행해 보세요:**

1. A 그룹과 B 그룹의 전환 데이터를 살펴봅니다.
2. 각 그룹의 전환율을 계산하고, 차이를 확인합니다.
3. 이 차이가 우연인지, 의미 있는 차이인지 생각해 봅시다.

```python
# 아래를 수행해 보세요:
# [문제 12] 여기에 코드를 작성해주세요.

np.random.seed(606)
group_a = np.random.binomial(1, 0.12, 1000)
group_b = np.random.binomial(1, 0.15, 1000)

conv_a = group_a.mean()
conv_b = group_b.mean()

print(f"A그룹 전환율: {conv_a*100:.2f}%")
print(f"B그룹 전환율: {conv_b*100:.2f}%")

# 차이: 0.20%p (B그룹이 A그룹보다 약간 높음)

# 이 차이는 우연인가, 의미 있는 차이인가?
from scipy import stats

success_a, fail_a = group_a.sum(), len(group_a)-group_a.sum()
success_b, fail_b = group_b.sum(), len(group_b)-group_b.sum()

table = np.array([[success_a, fail_a],[success_b, fail_b]])
chi2, p, dof, expected = stats.chi2_contingency(table)
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"p-value: {p:.4f}")

# 우연
```
